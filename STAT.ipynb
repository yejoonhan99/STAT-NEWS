```python
import pandas as pd
import numpy as np
import spacy
import en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
data = pd.read_csv('all-coi-recs-UTF8.csv')
data
```

```python
data['COIS'] = data['COIS'].astype('string')
data['COIS'].dtype
```

```python
data = data[data['COIS'].notna()] # Drop rows with missing COIS
data
```

```python
counts = data['COIS'].value_counts(ascending = False).to_frame()
counts
```

```python
counts = counts.reset_index()
```

```python
counts = counts.rename(columns = {'index' : 'COIS', 'COIS' : 'Counts'})
counts
```

```python
cois_more_than_once = counts[counts['Counts'] > 1] # COIS that is repeated more than once
cois_more_than_once # The output has 3502 rows.
```

```python
cois_once = counts[counts['Counts'] == 1]
cois_once
```

```python
cois_more_than_once['NER'] = cois_more_than_once['COIS'].apply(lambda sent: [ent.label_ for ent in nlp(sent).ents])
cois_more_than_once
```

```python
cois_more_than_once['Ent'] = cois_more_than_once['NER'].apply(lambda x : 'Ent' if len(x) > 0 else 'No Ent')
cois_more_than_once
```

```python
no_cois_more_than_once = cois_more_than_once[cois_more_than_once['Ent'] == 'No Ent']['COIS'].tolist()
no_cois_more_than_once
```

```python
cois_once['NER'] = cois_once['COIS'].apply(lambda sent: [ent.label_ for ent in nlp(sent).ents])
cois_once
```

```python
cois_once['Ent'] = cois_once['NER'].apply(lambda x : 'Ent' if len(x) > 0 else 'No Ent')
cois_once
```

```python
no_cois_once = cois_once[cois_once['Ent'] == 'No Ent']['COIS'].tolist()
no_cois_once
```

```python
joined_no_cois = no_cois_once + no_cois_more_than_once
```

```python
data = data.loc[~data['COIS'].isin(joined_no_cois)]
data
```

```python
data['NER'] = data['COIS'].apply(lambda sent: [ent.label_ for ent in nlp(sent).ents])
data
```

```python
data['Ent'] = data['NER'].apply(lambda x : 'Entity Exists' if len(x) > 0 else 'No Ent')
data
```

```python
data = data[data['Ent'] == 'Entity Exists']
data
```

```python
# For Tokenization (Step 6, 7, 8, 9)
```

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
tokenized = [nltk.sent_tokenize(coi) for coi in data['COIS'].tolist()]
```

```python
def ent_lab(doc):
    entity_labels = []
    if doc.ents:
        for ent in doc.ents:
            entity_labels.append(ent.label_)
    return entity_labels    
```

```python
data['Tokenized Sentence'] = tokenized
data
```

```python
data['is_entity'] = data['Tokenized Sentence'].apply(lambda x: [ent_lab(nlp(x[i])) for i in np.arange(len(x))])
data
```

```python
def is_empty(list_of_lists):
    if all([not sublist for sublist in list_of_lists]):
        return "Empty"
    else:
        return "Not empty"
```

```python
data['check_entity'] = data['is_entity'].apply(lambda x : is_empty(x))
data
```

```python
data = data[data['check_entity'] == 'Not empty']
data
```

```python
def entityfinder(raw_text):
    entities = []
    doc = nlp(raw_text)
    if doc.ents:
        for entity in doc.ents:
            entities += [entity.text]
    return entities
```

```python
def entityfinder_with_label(raw_text):
    entities = []
    doc = nlp(raw_text)
    if doc.ents:
        for entity in doc.ents:
            entities += [(entity.text, entity.label_)]
    return entities
```
